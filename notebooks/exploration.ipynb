```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. Load Data
df = pd.read_csv('../data/CO2_Emissions.csv')

# Drop text columns for the correlation map
numeric_df = df.select_dtypes(include=[np.number])

# 2. Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='RdYlGn', fmt=".2f")
plt.title('Correlation Matrix (CO2 Emissions Dataset)')
plt.show()

# ==========================================
# 3. Linear Regression Plot: Engine Size vs CO2 Emissions
# ==========================================
plt.figure(figsize=(8, 6))

sns.regplot(
    data=df, 
    x='Engine Size(L)', 
    y='CO2 Emissions(g/km)', 
    scatter_kws={'alpha': 0.4, 'color': 'green', 's': 15}, #Points
    line_kws={'color': 'purple', 'linewidth': 1.5}    #Line     
)

plt.title('Linear Algebra Model: Best Fit Line (Engine Size vs CO2)')
plt.xlabel('Engine Size (Liters)')
plt.ylabel('CO2 Emissions (g/km)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
